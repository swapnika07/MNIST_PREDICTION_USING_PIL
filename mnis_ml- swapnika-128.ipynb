```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os
from PIL import Image
```

```python
df = pd.read_csv("C:/Users/YASH/Desktop/mnist_data-29-4-2022/mnist_data1.csv")
```

```python
df.describe()
```

# Converting the dataset into X - as an input variable and y - Target  variable

```python
y=df['Target'] #target variable
X=df.iloc[:,0:784] #input variable
```

# split into train and test

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y, train_size = 0.75, random_state = 100)
```

```python
print(X_train.shape, y_train.shape)
print(X_test.shape, y_test.shape)
```

# b. Separating Categorical and Numerical Columns:

```python
X_train.dtypes
```

```python
X_train_num = X_train.select_dtypes(include=['float64'])
X_train_num.head()
```

```python
X_train_num.shape
```

# Standardizing the numerical values for train data.

```python
#Importing the standard
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()

X_train_num_rescaled = pd.DataFrame(scaler.fit_transform(X_train_num),
                                 columns = X_train_num.columns,
                                 index = X_train_num.index)

X_train_num_rescaled.head()
```

# DataPreparation for Test Data

```python
X_test.head()
```

```python
X_test.dtypes
```

```python
X_test_num = X_test.select_dtypes(include=['float64'])
X_test_num.head()
```

# Standardizing the numerical values for train data.


```python
X_test_num_rescaled = pd.DataFrame(scaler.transform(X_test_num),
                                 columns = X_test_num.columns,
                                 index = X_test_num.index)
```

```python
X_test_num_rescaled.head()
```

## As it is a Classification problem we are going to use Classification Task

```python
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression()
classifier.fit(X_train_num_rescaled, y_train)
```

```python
y_test_pred = classifier.predict(X_test_num_rescaled)
```

```python
temp_df = pd.DataFrame({'Actual' : y_test,'Predicted' :y_test_pred })
temp_df.head()
```

```python
from sklearn import metrics
metrics.accuracy_score(y_test, y_test_pred)
```

# Accuracy Rate Using Logistic Regression: 88%

# KNN CLASSIFIER


```python
from sklearn.neighbors import KNeighborsClassifier
neighbour = KNeighborsClassifier()
neighbour.fit(X_train_num_rescaled, y_train)
```

```python
y_test_pred = neighbour.predict(X_test_num_rescaled)
```

```python
temp_df = pd.DataFrame({'Actual' : y_test,'Predicted' :y_test_pred })
temp_df.head()
```

```python
from sklearn import metrics
metrics.accuracy_score(y_test, y_test_pred)
```

# Accuracy Rate Using KNN CLASSIFIER : 92%

# DECISION TREE CLASSIFIER

```python
from sklearn.tree import DecisionTreeClassifier
tree = DecisionTreeClassifier()
tree.fit(X_train_num_rescaled,y_train)
```

```python
y_test_pred = tree.predict(X_test_num_rescaled)
```

```python
temp_df = pd.DataFrame({'Actual' : y_test,'Predicted' :y_test_pred })
temp_df.head()
```

```python
from sklearn import metrics
metrics.accuracy_score(y_test, y_test_pred)
```

# Accuracy Rate Using Decision Tree Classifier : 77%

# Ensemble Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
rfc = RandomForestClassifier()
rfc.fit(X_train_num_rescaled,y_train)
```

```python
y_test_pred = rfc.predict(X_test_num_rescaled)
```

```python
temp_df = pd.DataFrame({'Actual' : y_test,'Predicted' :y_test_pred })
temp_df.head()
```

```python
from sklearn import metrics
metrics.accuracy_score(y_test, y_test_pred)
```

# Accuracy Rate Using Random Forest Accuracy Rate : 94%

```python
from sklearn.ensemble import GradientBoostingClassifier
gbc = GradientBoostingClassifier()
gbc.fit(X_train_num_rescaled,y_train)
```

```python
y_test_pred = gbc.predict(X_test_num_rescaled)
```
